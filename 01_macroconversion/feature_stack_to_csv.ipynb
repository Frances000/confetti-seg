```python
# From Labarchives page 8 - Create csv versions of feature stacks for machine learning
# I think the GUI processing in ImageJ is the primary source of the slowdown.
# This version achieves the same output with a speedup of at least 100x by writing the csv directly(hopefully)
```

```python
import imagej
import numpy as np
import pandas as pd

ij = imagej.init('sc.fiji:fiji')
# load feature stack with imageplus
imp = ij.io().open("TEMP PATH")
stack = ij.py.from_java(imp)
if stack.shape[0] < 20:
    reshaped = stack.reshape(stack.shape[0], -1).T
else:
    reshaped = stack.reshape(-1, stack.shape[-1])
# populate csv
df = pd.DataFrame(reshaped)
df.columns = [f"feat_{i}" for i in range(reshaped.shape[1])]
df.to_csv("output_feature_stack.csv", index=False)
# do not include data frame row labels in csv
```
